```python
import pandas as pd 
from moduls import DataPipeline, FeedFowardModel
```

5 year time horizont:

```python
# import the data
df_5 = pd.read_csv('data/data_5y_t.csv')

# creating the pipleline object
dataset_5y = DataPipeline(
    df = df_5, 
    label_col=  'overperf', 
    date_col = 'calendar_date', 
    batch_size = 256,
    test_from = '2014.01.01', 
    cols_to_drop=  ['ticker']
)
```

```python
# creating the modell
model_5y = FeedFowardModel(input_size = dataset_5y.feature_num,
                  hidden_size1 = 15,
                  num_classes = len(dataset_5y.y_encoder.classes_),
                  gpu= False
                )


# train the modell
model_5y.fit(epochs = 20,
          lr = 0.001,
          train_loader = dataset_5y.train_loader,
          val_loader = dataset_5y.test_loader)
```

```python
model_5y.evaluate(dataset_5y.test_loader)
```
